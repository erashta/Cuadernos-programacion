````c++
#include <iostream>
#include <functional>

int obtenerEntero()  //parte 1
{
    std::cout << "Ingresa un entero: ";
    int x{};
    std::cin >> x;
    return x;
}

char obtenerOperacion()
{
    char op{};

    do
    {
        std::cout << "Ingresa una operacion ('+', '-', '*', '/'): ";
        std::cin >> op;
    }
    while (op!='+' && op!='-' && op!='*' && op!='/');
         return op;
}

// parte 2
int sumar(int x, int y)
{
    return x + y;
}

int restar(int x, int y)
{
    return x - y;
}

int multiplicar(int x, int y)
{
    return x * y;
}

int dividir(int x, int y)
{
    return x / y;
}

//parte3
using funcionAritmetica = std::function<int(int, int)>;

//parte 4
funcionAritmetica obtenerfuncionAritmetica()(char op)
{
    switch (op)
    {
    case '+': return &sumar;
    case '-': return &restar;
    case '*': return &multiplicar;
    case '/': return &dividir;
    }

    return nullptr;
}

int main()
{
    int x{ obtenerEntero() };
    char op{ obtenerOperacion() };
    int y{ obtenerEntero() };

  funcionAritmetica fcn{obtenerfuncionAritmetica(op)};
    if (fcn)
        std::cout << x << ' ' << op << ' ' << y << " = " << fcn(x, y) << '\n';

    return 0;
}
````

**Ejercicio**

`Punteros y const`

Un uso común de punteros a elementos const es como parámetros de función. Sin esto, una función que toma un puntero como parámetro puede modificar el valor pasado como argumento. Declarar el parámetro como const evita que se modifique (accidentalmente o no).

````c++
// punteros como argumentos:
#include <iostream>
using namespace std;

void incrementa_todo(int* inicio, int* parar)
{
  int *actual = inicio;
  while (actual != parar) {
    ++(*actual);  //valor de incremento apuntado
    ++actual;     // puntero de incremento
  }
}

void imprimir_todo(const int* inicio, const int* parar)
{
  const int *actual = inicio;
  while (actual != parar) {
    cout << *actual << endl;
    ++actual;     // puntero de incremento
  }
}

int main()
{
  int numeros[] = {10,20,30};
  incrementa_todo(numeros,numeros+3);
  imprimir_todo(numeros,numeros+3);
  return 0;
}
````

El siguiente resultado es generado por este programa: `11 21 31`. 
    
En este ejemplo, `imprimir_todo` usa punteros que apuntan a valores constantes. Estos punteros no son constantes, por lo que los propios punteros aún pueden incrementarse o asignarse direcciones diferentes. Además de apuntar a valores constantes, también es posible declarar los punteros como constantes.

**Ejercicio**

¿Cuál es la diferencia entre un puntero constante y un puntero a constante?

Un puntero `const` es un puntero cuya dirección no se puede cambiar (no se puede volver a apuntar a algún otro objeto).

Un puntero a `const` es un puntero que apunta a un objeto `const` y, por lo tanto, el valor de ese objeto no se puede cambiar a través del puntero.
