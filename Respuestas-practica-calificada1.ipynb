**Versión 1**

```c++
#include <iostream>
using namespace std;

void dividir_numero(int num, int resultado[],int &n)
{
	if (num > 0)
	{
		dividir_numero(num/10, resultado,n);
		resultado[n++]= (num % 10);
	}
}

void agregar(int a[], int b[], int na, int nb, int resultado[], int &n)
{
	int i = 0;
	while (i < na && i < nb)
	{
		int sum = a[i] + b[i];

		dividir_numero(sum, resultado,n);

		i++;
	}
	while (i < na) {
		dividir_numero(a[i++], resultado,n);
	}

	while (i < nb) {
		dividir_numero(b[i++], resultado,n);
	}
}

int main()
{
	int a[] = { 23, 5, 2, 7, 87 };
	int b[] = { 4, 67, 2, 8 };
	
	int resultado[100];
	
	int n=0;	
	int na = sizeof(a)/sizeof(a[0]);
	int nb = sizeof(b)/sizeof(b[0]);
	
	agregar(a, b, na, nb, resultado,n);

	for (int i=0;i<n;i++) {
		cout << resultado[i] << " ";	
	}	
}
```

```c++
#include <iostream>
#include <vector>
using namespace std;

void dividir_numero(int num, vector<int> &resultado)
{
	if (num > 0)
	{
		dividir_numero(num/10, resultado);
		resultado.push_back(num % 10);
	}
}

void agregar(vector<int> const &a, vector<int> const &b, vector<int> &resultado)
{
	int m = a.size(), n = b.size();

	int i = 0;
	while (i < m && i < n)
	{
		int sum = a[i] + b[i];

		dividir_numero(sum, resultado);

		i++;
	}

	while (i < m) {
		dividir_numero(a[i++], resultado);
	}

	while (i < n) {
		dividir_numero(b[i++], resultado);
	}
}

int main()
{

	vector<int> a = { 23, 5, 2, 7, 87 };
	vector<int> b = { 4, 67, 2, 8 };

	
	vector<int> resultado;
	agregar(a, b, resultado);

	for (int i: resultado) {
		cout << i << " ";
	}

	return 0;
}
```

```c++
#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;


vector<int> encuentraKElementosCercanos(vector<int> const &entrada, int objetivo, int k)
{
	int i = lower_bound(entrada.begin(), entrada.end(), objetivo) - entrada.begin();

	int izquierda = i - 1;
	int derecha = i;

	while (k-- > 0)
	{
		
		if (izquierda < 0 || (derecha < entrada.size() &&
				abs(entrada[izquierda] - objetivo) > abs(entrada[derecha] - objetivo))) {
			derecha++;
		}
		else {
			izquierda--;
		}
	}

	return vector<int>(entrada.begin() + izquierda + 1, entrada.begin() + derecha);
}

int main()
{
	vector<int> entrada = { 10, 12, 15, 17, 18, 20, 25 };
	int objetivo = 16, k = 4;

	vector<int> result = encuentraKElementosCercanos(entrada, objetivo, k);
	for (int i: result) {
		cout << i << " ";
	}

	return 0;
}
```

```c++
#include <iostream>
using namespace std;

int lower_bound(int entrada[],int n,int value) {
    int first=0;
	int count = n;
	int it, step;
	while (count > 0)
    {
        it = first; 
        step = count / 2; 
        it += step;
 
        if (entrada[it] < value)
        {
            first = ++it; 
            count -= step + 1; 
        }
        else
            count = step;
    }
 
    return first;
}

void encuentraKElementosCercanos(int entrada[],int n, int objetivo, int k,int &izquierda,int &derecha){			
	int i = lower_bound(entrada,n,objetivo);

	izquierda = i - 1;
	derecha = i;

	while (k-- > 0)
	{		
		if (izquierda < 0 || (derecha < n &&
				abs(entrada[izquierda] - objetivo) > abs(entrada[derecha] - objetivo))) {
			derecha++;
		}
		else {
			izquierda--;
		}
	}
	izquierda++;	
}

int main()
{
	int entrada[] = { 10, 12, 15, 17, 18, 20, 25 };
	int n;
	int objetivo = 16, k = 4;
    int izquierda, derecha;
    n = sizeof(entrada)/sizeof(entrada[0]);
	encuentraKElementosCercanos(entrada, n, objetivo, k, izquierda, derecha);
	for (int i=izquierda;i<derecha;i++) {
		cout << entrada[i] << " ";
	}	
}
```

```c++
#include <iostream>
#include <cstdlib> //para rand()
#include <ctime> //para time()
using namespace std;

void bubbleSort(int arr[], int n) {
    for(int i = 0; i < n - 1; i++) {
        for(int j = 0; j < n - i - 1; j++) {
            if(arr[j] > arr[j+1]) {
                int temp = arr[j];
                arr[j] = arr[j+1];
                arr[j+1] = temp;
            }
        }
    }
}

int main() {
    int arr[100];
    int n, k;
    cout << "Ingrese la longitud n del arreglo: ";
    cin >> n;
    //generar el arreglo aleatoriamente
    srand(time(NULL)); //semilla aleatoria
    for(int i = 0; i < n; i++) {
        arr[i] = rand() % 100; //generar numero aleatorio entre 0 y 99
    }
    
    cout << "Arreglo original: ";
    for(int i = 0; i < n; i++) {
        cout << arr[i] << " ";
    }
    cout << endl;
    
    cout << "Ingrese la posicion k: ";
    cin >> k;
    bubbleSort(arr, n);
    cout << "Arreglo ordenado: ";
    for(int i = 0; i < n; i++) {
        cout << arr[i] << " ";
    }
    cout << endl;
    
    cout << "El elemento en la posicion " << k << " es: " << arr[k-1] << endl;
    return 0;
}
```

```c++
// el programa muestra un equipo con formacion 4 - 3 - 3
// 4 en la defensa, 3 en el medio campo y 3 delanteros

#include <iostream>
#include <stdlib.h>
#include <time.h>
using namespace std; 
int const N=11, M=3;

void quicksort(int[], int[], int , int);
int particionar(int[], int[], int, int);

int main(){
	int PLANTEL[N][M];
	int JUGADOR[N]; 
	int FORMACION[]={4,3,3};
	char TIPO[3][20]={"Defensa", "Mediocampo","Delantero"};
	int R[N]; 
    int num, c;
    srand(time(NULL));
    // se  mide la habilidad de 100 jugadores de futbol 
	// en la escala de 1 a 5
    // en la defensa, mediocampo y como delantero
    for(int i=0;i<N;i++)    {
    	for(int j=0;j<M;j++)    {
	        PLANTEL[i][j]= 1 + rand() % (5);        
    	}
    }
    cout<<"\tTABLA DE RENDIMIENTO\n";
    cout<<"\tJUG.\t"<<"DEFEN.\t"<<"MEDIO.\t"<<"DELANT."<<endl;
    for(int i=0;i<N;i++)    {
    	cout<<"\t"<<i+1<<"\t";
    	for(int j=0;j<M;j++)    {
	        cout<<PLANTEL[i][j]<<"\t";
    	}
    	cout<<endl;
    }
    
    for(int j=0;j<M;j++)    {    	
    	for(int i=0;i<N;i++)    {
    		R[i]=PLANTEL[i][j];
    		JUGADOR[i]=i;
    	}
    	quicksort(R,JUGADOR,0,N-1);
    	
    	int nJugadores=FORMACION[j];// numero de jugadores a seleccionar
    	cout<<"\tJugadores de "<<TIPO[j]<<endl;
    	for(int i=0;i<nJugadores;i++){
    		cout<<"\t"<<JUGADOR[i]+1<<"\t"<<R[i]<<endl;
        	// ponemos ceros al rendimiento de los jugadores seleccionados
        	for(int k=0;k<M;k++){
        		PLANTEL[JUGADOR[i]][k]=0;
			}    		
		}
	}
    return 0;
}

void quicksort(int a[], int b[],int low, int high){
    if (low<high){
        //particionar desde a[low] hasta a[high]
        //y devolver el indice del pivote
        int pivotIdx=particionar(a,b,low,high);

        //recursivamente ordenar las dos partes
        quicksort(a,b,low,pivotIdx-1);
        quicksort(a,b,pivotIdx+1,high);
    }
}

int particionar(int a[],int b[], int i, int j){
    float p=a[i]; // p es el pivote
    // Las sublistas S1, S2, (Antes y despues del pivote)
    // estan vacias al inicio
    int m = i;
    // pasamos por cada elemento en la
    // region no particionada
    for (int k=i+1; k<=j; k++){
        if (a[k]>p)
        {
            m++;
            swap(a[k],a[m]);
            swap(b[k],b[m]);
        }
        else{
            // no se hace nada
        }
    } 
    // intercambiar el pivote con a[m]
    swap(a[i],a[m]);
    swap(b[i],b[m]);    
    //retornamos el indice del pivote
    return m;
}
```
