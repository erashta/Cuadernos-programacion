**Ejercicio**


Defin el sinónimo `Arr` de un arreglo de enteros de capacidad 20. Genere aleatoriamente un arreglo de notas, las notas son números enteros del 0 al 20, y muestre dicho arreglo. Luego defina recursivamente las funciones `ordenamiento_de_burbuja_indices_pares` y `ordenamiento_de_burbuja_indices_impares` que ordena descendentemente y ascendentemente -según el método de ordenamiento de burbuja- los elementos de índice par e impar, respectivamente, de arreglos del tipo `Arr`.

Finalmente, utilice dichas funciones para ordenar los elementos de índice par descendentemente y los elementos de índice impar ascendentemente del arreglo de notas generado y vuelva a mostrar dicho arreglo.

La salida sería:

```

	Notas: 1 4 9 19 8 10 10 9 15 10 2 19 20 4 20 7 3 15 16 16

	Notas: 20 4 20 4 16 7 15 9 10 10 9 10 8 15 3 16 2 19 1 19
``` 
Sugerencia: La palabra clave typedef le permite al programador crear nuevos nombres para tipos como int o, más comúnmente en C ++, tipos con plantilla, literalmente significa "definición de tipo", typedef se puede usar tanto para proporcionar más claridad a su código como para facilitar la realización de cambios en los tipos de datos subyacentes que utiliza.


````c++
#include <iostream>
#include <iomanip>
using namespace std;
  
void swap(int *xp, int *yp){
    int temp = *xp;    *xp = *yp;    *yp = temp; } // uso de punteros
    
void ordenamiento_de_burbuja_indices_pares(int arr[], int n){
    // Caso base
    if (n == 1)
        return;
  
    int count = 0;
    for (int i=0; i<n-1; i+=2)
        if (arr[i] < arr[i+2]){
            swap(&arr[i], &arr[i+2]);
            count++;
        }
  
    if (count==0)
           return;
  
    ordenamiento_de_burbuja_indices_pares(arr, n-2);
}
void ordenamiento_de_burbuja_indices_impares(int arr[], int n){
    // Caso base
    if (n == 1)
        return;
  
    int count = 0;
    for (int i=1; i<n-1; i+=2)
        if (arr[i] > arr[i+2]){
            swap(&arr[i], &arr[i+2]);
            count++;
        }
  
    if (count==0)
           return;
  
    ordenamiento_de_burbuja_indices_impares(arr, n-2);
}
  

void printArray(int arr[], int n){
    for (int i=0; i < n; i++)
        cout<<setw(3)<<arr[i];
    cout<<endl;
}

int main(){
    int arr[] = {1, 4, 9, 19, 8, 10, 10, 9, 15, 10, 2, 19,
    20, 4, 20, 7, 3, 15, 16, 16};
    int n = sizeof(arr)/sizeof(arr[0]);
    ordenamiento_de_burbuja_indices_pares(arr, n);
    ordenamiento_de_burbuja_indices_impares(arr, n);
    cout<<"Sorted array : "<<endl;
    printArray(arr, n);
    return 0;
}
````

````c++
#include<iostream>
#include<iomanip>
#include <cstdlib>

using namespace std;

const int nalumnos=4;  
const int ncursos=5; 

void mostrar_notas(int notas[][nalumnos]){
    for (int i=0;i<ncursos;i++){
        for (int j=0;j<nalumnos;j++)
             cout<<setw(5)<<notas[i][j];
        cout<<endl;
    }
}
void swap(float A[], int i, int j){
    float aux; aux = A[i]; A[i]=A[j]; A[j]=aux;
}
void swap(int A[], int i, int j){
    int aux; aux = A[i]; A[i]=A[j]; A[j]=aux;
}
int  main() { 
    int   notas[ncursos][nalumnos]={0};
    float promedio_alumnos[nalumnos]={0};
    int   codigo_alumnos[nalumnos]={0};
    
    
    for (int i=0;i<ncursos;i++)
        for (int j=0;j<nalumnos;j++)
            notas[i][j] = 1+rand()%20;
    cout<<"Notas "<<endl;        
    mostrar_notas(notas);
    
    for (int j=0;j<nalumnos;j++){
        float promedio    = 0;
        codigo_alumnos[j] = j+1;
        for (int i=0;i<ncursos;i++){  
            promedio+=notas[i][j] ;
            promedio_alumnos[j]=promedio/5.0;  
        }
    }
    for (int i =0;i<nalumnos-1;i++)
        for (int j =0;j<nalumnos-1;j++)
            if (promedio_alumnos[j]>promedio_alumnos[j+1]){
                swap(promedio_alumnos,i,j);
                swap(codigo_alumnos,i,j);
            }
    cout<<setw(6)<<"Codigo"<<setw(10)<<"Promedio"<<endl;        
    for (int i =0;i<nalumnos;i++)
        cout<<setw(5)<<codigo_alumnos[i]<<setw(7)<<promedio_alumnos[i]<<endl;
}
````

**Ejercicio**


Sea L un arreglo de registros que contiene el código de estudiante, código de curso y nota final obtenida por el estudiante en ese curso. Para cada estudiante pueden existir varios nodos:

a) Halla el promedio del estudiante con código X.

b) Crea P, un arreglo de registros que contenga el código de estudiante y su promedio sobre todos los cursos que figuran en el arreglo L.

c) Ordena los elementos de P respecto al código de estudiante.

d) Ordena los elementos de P respecto al promedio.



````c++
#include <iostream>
using namespace std; 
struct info {
 int codEst;
 int codCurso;
 int notaFinal;
};
struct infoEst {
 int codigo; 
 float promedio;
};
float calculaPromedio(info L[], int k, int x){
    float prom=0;
    int numCursos=0;
    for (int i = 0; i < k; i++) {
        if (L[i].codEst==x){
            prom+=L[i].notaFinal; numCursos++;
        }
    } 
    return prom/numCursos;
}
int buscarCodigo(infoEst P[], int m, int cod){
    for (int i = 0; i < m; i++) {
        if (P[i].codigo==cod)
                return 1;
    }
    return 0;
}
int Particion(infoEst A[],int p,int r){
    float x = A[r].promedio;
    int i = p-1;
    for (int j = p ; j<=r-1;j++){
        if (A[j].promedio <= x){
            i++;
            infoEst aux = A[i]; A[i]=A[j];
            A[j] = aux;
        }
    }
    infoEst aux =A[i + 1]; A[i+1]=A[r];
    A[r]=aux;
    return i + 1;

}
void Ordena(infoEst A[],int p,int r){
    while (p < r){
        int q = Particion(A,p,r);
        Ordena(A,p,q-1);
    p = q + 1;
    }
}
int main(){
    info L[20] = { {1,1,16},{1,2,18},{2,1,15},{3,2,20}};
    infoEst P[20] = { {0 ,0.} };    
    int m=0,k=4;
    for (int i = 0; i < k; i++) {
        int cod = L[i].codEst;
        if (!buscarCodigo(P, m, cod)){
            float promedio = calculaPromedio(L, k, cod);
            P[m].codigo = cod;
            P[m].promedio = promedio;
            m++;
        } 
    }
    Ordena(P,0,m-1);
     for (int i = 0; i < m; i++) {
        cout<<" "<<P[i].codigo<<"  "<<P[i].promedio<<endl;
    } 
    cout<<endl;
    return 0;
}
````

**Ejercicio**

Considera un arreglo ordenado con elementos repetidos, escribe un programa que inserte un elemento en el arreglo de tal forma que si elemento ya se encuentra presente lo inserta después de todas las repeticiones.


````c++
#include <iostream>
using namespace std; 
void insercionVectorOrdenado(int A[], int n, int x){
    int ipos = n;
    while(ipos && (x<A[ipos-1])){
      A[ipos] = A[ipos-1];
        ipos--;
    }
    A[ipos]= x;
} 
void printArray(int[],int);
int main(){
    int A[50] = { 0, 1, 1 , 1, 3, 3, 4, 7, 7, 7, 9 };
    int n = 11;
    int x=4;
    insercionVectorOrdenado(A,n,x);   
    printArray(A,n+1);
    return 0;
}
void printArray(int arr[], int n)
{
    for (int i = 0; i < n; i++) {
        cout<<" "<<arr[i];
    } 
    cout<<endl;
}
````

**Ejercicio**

Invierte el orden de un arreglo ordenado mientras recorre el arreglo.


````c++
#include <iostream>
using namespace std; 
void invertirOrden(int A[], int n){
    for (int i = 0; i < n/2; i++) {
        int aux = A[i];
        A[i]=A[n-i-1];A[n-i-1]=aux;
    } 
} 
void printArray(int[],int);
int main(){
    int A[50] = { 0, 1, 1 , 1, 3, 3, 4, 7, 7, 7, 9, 11 };
    int n = 12;    
    invertirOrden(A,n);   
    printArray(A,n);
    return 0;
}
void printArray(int arr[], int n)
{
    for (int i = 0; i < n; i++) {
        cout<<" "<<arr[i];
    } 
    cout<<endl;
}
````
