 ¿Es `new`  lo mismo que `malloc()` y lo mismo es `delete` que `free()`?
 
No. Son similares, no iguales. `malloc()` simplemente asigna memoria y `free()` lo libera. En contra de esto, `new` asigna memoria y llama al constructor, mientras que `delete` llama a destructor y desasigna memoria. La llamada al constructor garantiza que el objeto se inicialice correctamente, mientras que la llamada al destructor garantiza que el objeto se destruya correctamente sin causar pérdidas de memoria. 

`malloc()` y `free()` deben usarse en un programa C porque en C no existe el concepto de objetos, constructores o destructores. `new` y `delete` debe usarse en un programa C++. 

Ten en cuenta que `malloc()` y `free()` no son compatibles con `new` y `delete`. Esto significa no asignar usando `malloc()` y liberar usando `delete` o no asignar usando `new` y liberar usando `free()`.

````c++
#include<iostream>
#include<cstdlib>
#include<ctime>

using namespace std;

int** generar(int,int);
void mostrar(int **,int,int);
int** transponer(int **,int,int);
void liberar(int **,int);


int main(){
    int **d_a,**d_t,f,c;
    cout<<"Ingresa el numero de filas: ";cin>>f;
    cout<<"Ingresa el numero de columnas: ";cin>>c;
    d_a=generar(f,c);
    
    cout<<"Esta es la matriz aleatoria: "<<endl;
    mostrar(d_a,f,c);
    d_t=transponer(d_a,f,c);
    
    cout<<"Esta es su matriz transpuesta: "<<endl;
    mostrar(d_t,c,f);
    
    liberar(d_t,c);
    liberar(d_a,f);
    
    cout<<endl;
    return 0;
}

int** transponer(int **d_a,int f,int c){
    int **transpuesta;
    transpuesta=generar(c,f);
    for(int i=0;i<f;i++){
        for(int j=0;j<c;j++){
            *(*(transpuesta+j)+i)=*(*(d_a+i)+j);
        }
    }
    return transpuesta;
    liberar(transpuesta,c);
}


void liberar(int **d_a,int f){
    for(int i=0;i<f;i++){
        delete d_a[i];
    }    
    delete d_a;
}

void mostrar(int **d_a,int f,int c){
    for(int i=0;i<f;i++){
        for(int j=0;j<c;j++){
            cout<<*(*(d_a+i)+j)<<"\t";
        }
        cout<<endl;
    }

}

int** generar(int f,int c){
    int **d_a;
    d_a=new int *[f];
    for(int i=0;i<f;i++){
        d_a[i]=new int [c];
    }
    srand(time(NULL));
    for(int i=0;i<f;i++){
        for(int j=0;j<c;j++){
            *(*(d_a+i)+j)=rand()%(10);
        }
    }
    return d_a;
        
}
````

Escribe un programa para determinar si los enteros en un arreglo siguen una progresión aritmética o geométrica. Una progresión aritmética es aquella en la que la diferencia entre cada par sucesivo de enteros es constante, por ejemplo, en el arreglo `[2, 4, 6, 8]` la diferencia entre los enteros es siempre 2. 

Una progresión geométrica es aquella en la que cada elemento del arreglo  es el producto del entero anterior multiplicado por alguna constante o razón, por ejemplo en el arreglo `[3, 9, 27, 81]` cada elemento es el resultado del elemento anterior multiplicado por 3.


````c++
#include <iostream>
using namespace std ;
bool esPA ( int*, int ) ;
bool esPG ( int*, int ) ;
const int MAX = 5 ;
int main( )

{
   int arr[ MAX ] ;
   bool pa, pg ;
    cout << "Ingresa " << MAX << " los terminos de la secuencia: " << endl ;
   for ( int i = 0 ; i < MAX ; i++ )
      cin >> arr[ i ] ;
   pA = esPA ( arr, MAX ) ;

    if (pa == true )
      cout << "La secuencia es un PA" << endl ;
    else
        cout << "La secuencia no es un PA" << endl ;

    pg = esPG ( arr, MAX ) ;
    if (pg == true )
        cout << "La secuencia  es un PG" << endl ;
    else
        cout << "La secuencia no es un GP" << endl ;
    return 0 ;
}
bool esPA ( int* arr, int size )
{
    int diff ;
    bool pa = true ;
    diff = arr[ 1 ] - arr[ 0 ] ;
    for ( int i = 2 ; i < size ; i++ )
    {
        int t = arr[ i ] - arr[ i - 1 ] ;
    if ( t != diff )
    {
        pa = false ;
        break ;
    }
}

    return pa ;
}
bool esGP ( int* arr, int size )
{
    float radio ;
    bool pg = true ;
    for ( int i = 0 ; i < size ; i++ )
    {
    if ( arr[ i ] == 0 )
    {
        pg = false ;
        break ;
        }
   }
    if ( pg == true )
    {
        radio = ( float ) arr[ 1 ] / ( float ) arr[ 0 ] ;
        for ( int i = 2 ; i < size ; i++ )
        {
            float t = ( float ) arr[ i ] / ( float ) arr[ i - 1 ] ;
            if ( t != radio )
            {   
                pg = false ;
                break ; 
                }
            }
    }
    return pg ;
}
````

Por último, desasignar la memoria. Nuevamente, verifica que haya desasignado la memoria correctamente usando un bucle for como en el ejercicio anterior.


````c++
#include <iostream>

int main()
{
    for (long int k=0; k<1000000000; k++)
    {
       
        double **A, **B, **C;
        A = new double*[2];
        B = new double*[2];
        C = new double*[2];

        for (int j=0; j<2; j++)
        {
            A[j] = new double[2];
            B[j] = new double[2];
            C[j] = new double[2];
        }

        for (int i=0; i<2; i++)
        {
            for (int j=0; j<2; j++)
            {
                A[i][j] = (i + j) * 1.0;
                B[i][j] = (i + j) * 10.0;
            }
        }

        // Calculamos C = A + B
        for (int i=0; i<2; i++)
        {
            for (int j=0; j<2; j++)
            {
                C[i][j] = A[i][j] + B[i][j];
            }
        }
        for (int j=0; j<2; j++)
        {
            delete[] A[j];
            delete[] B[j];
            delete[] C[j];
        }

        delete[] A;
        delete[] B;
        delete[] C;
    }

    return 0;
}
````
