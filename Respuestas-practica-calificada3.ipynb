### Ejercicio 1

```c++
#include <iostream>
using namespace std;
int esta_en_el_arreglo(int *p, int n,int val){
    for(int i=0;i<n;i++){
        if (*p==val)
            return 1;
        else
            p++;
    }
    return 0;
}
int main(){
    int i, n=0, x=1;
    int max=-1,imax=-1;
    int *p = nullptr;
    while (x>0){
        cout<<"Ingrese un valor entero positivo (0 para salir) : ";
        cin>>x;
        if ( (x>0) && !esta_en_el_arreglo(p, n, x) ) {
            int *aux =new int[n+1];
            if (p!=nullptr){
                if (x>max){// insertamos al final
                    for (i=0;i<n;i++)	aux[i]=p[i];           	
                }
                else { // insertamos al inicio
                    for (i=n;i>0;i--)	aux[i]=p[i-1];
                }
                delete[] p;
            }
            aux[i]=x;
                p=aux;
            if (x>max) max=x; //actualizamos el valor del maximo
            n++; //actualizamos la longitud del arreglo	
        }
    }
    cout<<"Arreglo formado: "<<endl;
    for(i=0;i<n;i++)
        cout<<p[i]<<" - ";
    cout<<endl<<"El valor maximo es: "<<max<<endl;
}
```

### Ejercicio 2

```c++
#include <iostream>

double** asignarMatriz(int filas, int cols) {
    double** matriz = new double*[filas];
    for (int i = 0; i < filas; i++) {
        matriz[i] = new double[cols];
    }
    return matriz;
}


void deasignarMatriz(double** matriz, int filas) {
    for (int i = 0; i < filas; i++) {
        delete[] matriz[i];
    }
    delete[] matriz;
}

int main() {
    double** A = asignarMatriz(2, 2);
    double** B = asignarMatriz(2, 2);
    double** C = asignarMatriz(2, 2);

    // 
    A[0][0] = 1.0;
    A[0][1] = 2.0;
    A[1][0] = 3.0;
    A[1][1] = 4.0;

    B[0][0] = 5.0;
    B[0][1] = 6.0;
    B[1][0] = 7.0;
    B[1][1] = 8.0;

    // 
    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 2; j++) {
            C[i][j] = A[i][j] + B[i][j];
        }
    }

    // 
    std::cout << "C = A + B:\n";
    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 2; j++) {
            std::cout << C[i][j] << " ";
        }
        std::cout << "\n";
    }

    // 
    deasignarMatriz(A, 2);
    deasignarMatriz(B, 2);
    deasignarMatriz(C, 2);

    return 0;
}
```

### Ejercicio 3

```c++
#include <iostream>
#include <cmath>
using namespace std;
struct Punto {
    double x;
    double y;
    double z;
};
int main() {
    Punto punto1, punto2;
    cout << "Ingrese las coordenadas del primer punto:\n";
    cout << "X: ";
    cin >> punto1.x;
    cout << "Y: ";
    cin >> punto1.y;
    cout << "Z: ";
    cin >> punto1.z;
    cout << "Ingrese las coordenadas del segundo punto:\n";
    cout << "X: ";
    cin >> punto2.x;
    cout << "Y: ";
    cin >> punto2.y;
    cout << "Z: ";
    cin >> punto2.z;
    double dx = punto2.x - punto1.x;
    double dy = punto2.y - punto1.y;
    double dz = punto2.z - punto1.z;
    double distancia = sqrt(dx * dx + dy * dy + dz * dz);
    cout << "La distancia entre los dos puntos es: " << distancia << endl;
    return 0;
}
```
