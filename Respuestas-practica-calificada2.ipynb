### Ejercicio 1

```c++
#include <iostream>
using namespace std;
void ordBurbuja(int* A, int n) {
    for (int i = 0; i < n - 1; i++) {
        for (int j = 0; j < n - i - 1; j++) {
            if (*(A + j) > *(A + j + 1)) {
                int temp = *(A + j);
                *(A + j) = *(A + j + 1);
                *(A + j + 1) = temp;
            }
        }
    }
}
int main() {
    int A[] = { 5, 2, 9, 4, 1, 8, 5, 6, 3, 0, 7 };
    int n = sizeof(A) / sizeof(int);
    ordBurbuja(A, n);
    std::cout << "Vector despues de ser ordenado: ";
    for (int i = 0; i < n; i++) {
        cout << *(A+i) << " ";
    }
    return 0;
}
```

```c++
#include <iostream>
#include <string>
using namespace std;

unsigned int strlen(const char * p)
{
    int i = 0;
    while (*(p + i) != '\0')
   	 i++;
    return i;
}


void rev(char *a)
{
    int n = strlen(a);
    for (auto i = 0; i < n/2; i++)
     swap(*(a+i), *(a + n-i-1));

}

int main()
{
    char a[10] = {  };
    cin >> a;
    rev(a);
    for (auto i :a)
      cout << i;
    return 0;
}
```

### Ejercicio 3

```c++
#include <iostream>
using namespace std;
void imprime(int* p, int n){
    cout<<"\np(x)="<<*p++;
    for(int i=1;i<n;i++){
        cout<<((*p<0)?"-":"+")<<abs(*p)<<"x^"<<i;
        p++;
    }
    cout<<endl;
}
void derivada(int* p, int n, int k){
    int t;
    for(int i=0;i<n;i++){
        t=i+1;	
        if (i<n-k){
            for(int j=2;j<k+1;j++) 
                t= t*(i+j);
        *p = *(p+k) * t ;
        }
        else
            *p=0;
        p++;
    }
}
int main(){
    int A[100];
    int k,n;
    cout<<"Ingrese grado del polinomio="; cin>>n;
    for (int i=0;i<n+1;i++){
        cout<<"A["<<i<<"]="; cin>>A[i];
    }
    imprime(A,n+1);	
    cout<<"\nIngrese k="; cin>>k;
    derivada(A,n+1,k);
    cout<<"\nLa derivada k-esima es \n";
    imprime(A,n-k+1);
}
```

```c++
#include <iostream>
using namespace std;
int const maxcadena=40;

int EstaBienEscrito(char *p){
    bool esBlanco=true;
    bool malEscrito=false;
    int pos=0;
    char palabra[maxcadena];
    while (*p){
        if  (*p==' ') {
            if (esBlanco)
                malEscrito=true;
            else {
                esBlanco=true;
    }
    } else {
            if ((*p>='a') && (*p<='z')){
        if (esBlanco)
            malEscrito=true;
        }
        else if  ((*p>='A') && (*p<='Z'))  {
        if (!esBlanco)
            malEscrito=true;
            }
            else if ((*p<'0') || (*p>'9')) {
                malEscrito=true;
            }
            esBlanco=false;;
        }
        if (malEscrito)
            return pos;
        else {
            p++; pos++;
        }
    }
    if (esBlanco)
         return pos-1;
    else
        return -1;
}
int main(){
    char cadena[maxcadena]="Hoy Es MiercolEs 8 De Mayo De 2023";
    cout<<"Ingrese una cadena: ";
    cin.getline(cadena,maxcadena);
    
    char *p = cadena;
    int pos = EstaBienEscrito(p);
    
    if (pos>-1){
        cout<<" Error en la posicion "<<pos<<endl;
        cout<<" "<<cadena<<endl;
        for (int i=0;i<pos+1;i++) cout<<".";
        cout<<"|"<<endl;
    }
    else{
        cout<<" "<<cadena<<endl;
        cout<<" Titulo bien escrito"<<endl;	
    }
}
```
